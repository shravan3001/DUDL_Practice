## Entropy and Cross-Entropy

```python
p = 0.25
q = 1 - p

# Binary cross entropy
H = - (p* np.log(p) + (1-p)*np.log((1-p)))

print(H)
```

```python
x = [0.25, 0.75]
H = 0
for p in x:
    H -= p*np.log(p)
print(H)
```

```python
# cross entropy

p = [1,0]
q = [0.25,0.75]

H = 0
for i in range(len(p)):
    H -= p[i] * np.log(q[i])

print(H)
```

```python
import torch.nn.functional as F
```

```python
q_tensor = torch.Tensor(q)
p_tensor = torch.Tensor(p)

F.binary_cross_entropy(q_tensor,p_tensor)
```

## Mean and Variance

```python
x = [1, 2, 4, 6, 5, 4, 0]
n = len(x)
mean1 = np.mean(x)
mean2 = np.sum(x) / n
print(f'{mean1=} {mean2=}')
```

```python
var1 = np.var(x)
var2 = (1/(n-1)) * np.sum((x-mean1)**2)
print(f'{var1=} {var2=}')
```

```python
var1 = np.var(x, ddof=1) # the 1 from (n-1) is coming from degrees of freedom and is called unbiased variance
var2 = (1/(n-1)) * np.sum((x-mean1)**2) # if u use n instead of n-1 you get biased variance
print(f'{var1=} {var2=}')
```

## T-Test using scipy

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
n = [30, 40] # samples in each dataset
mu = [1, 2]  # population mean in each dataset
dc = len(n)
data = []
for i in range(dc):
    d = mu[i] + np.random.randn(n[i])
    data.append(d)
print(data)
```

```python
col = ['ro', 'bs']
fn = [np.zeros, np.ones]
for i in range(dc):
    plt.plot(fn[i](n[i]), data[i], col[i], markerfacecolor='w', markersize=14)
plt.xlim([-1, 2])
plt.xticks([0,1], labels=['Group 1', 'Group 2'])
plt.show()
```

```python
# _ind = independent samples
t, p = stats.ttest_ind(data[0], data[1])
print(t)
print(p)
```

## Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
display.set_matplotlib_formats('svg')
```

```python
def fx(x):
    return 3*x**2 - 3*x + 4

def deriv(x):
    return 6*x - 3
```

```python
# plot function and derivative

x = np.linspace(-2, 2, 2001)

plt.plot(x, fx(x), x, deriv(x))
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y', 'dy'])
plt.show()
```

```python
# random starting point
localmin = np.random.choice(x, 1)

# learning param
learning_rate = 0.01
training_epochs = 100

# run through the training
for i in range(training_epochs):
    grad = deriv(localmin)
    localmin = localmin - grad * learning_rate

print(localmin)
```

```python
plt.plot(x, fx(x), x, deriv(x))
plt.plot(localmin, deriv(localmin), 'ro')
plt.plot(localmin, fx(localmin), 'ro')

plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)', 'df', 'f(x) min'])
plt.title(f'Emperical local minimum: {localmin[0]}')
plt.show()
```

```python
# random starting point
localmin = np.random.choice(x,1)

# learning parameters
learning_rate = .01
training_epochs = 100

# run through training and store all the results
modelparams = np.zeros((training_epochs,2))
for i in range(training_epochs):
  grad = deriv(localmin)
  localmin = localmin - learning_rate*grad
  modelparams[i,0] = localmin
  modelparams[i,1] = grad
```

```python
# plot the gradient over iterations

fig,ax = plt.subplots(1,2,figsize=(12,4))

for i in range(2):
  ax[i].plot(modelparams[:,i],'o-')
  ax[i].set_xlabel('Iteration')
  ax[i].set_title(f'Final estimated minimum: {localmin[0]:.5f}')

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')

plt.show()
```

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

def fx(x):
  return np.cos(2*np.pi*x) + x**2

def deriv(x):
  return -2 * np.pi * np.sin(2*np.pi*x) + 2*x

# initialize x
x = np.linspace(-2, 2, 2001)

# random starting point
localmin = np.random.choice(x,1)
#localmin = 0

# learning parameters
learning_rate = .01
training_epochs = 100

# run through training and store all the results
modelparams = np.zeros((training_epochs,2))
for i in range(training_epochs):
  grad = deriv(localmin)
  localmin = localmin - learning_rate*grad
  modelparams[i,0] = localmin
  modelparams[i,1] = grad
```

```python
# plot the gradient over iterations

fig,ax = plt.subplots(1,2,figsize=(12,4))

for i in range(2):
  ax[i].plot(modelparams[:,i],'o-')
  ax[i].set_xlabel('Iteration')
  ax[i].set_title(f'Final estimated minimum: {localmin[0]:.5f}')

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')

plt.show()
```

## GD in 2D

```python
# import all necessary modules
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym # sympy to compute the partial derivatives

import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
# the "peaks" function
def peaks(x,y):
  # expand to a 2D mesh
  x,y = np.meshgrid(x,y)
  
  z = 3*(1-x)**2 * np.exp(-(x**2) - (y+1)**2) \
      - 10*(x/5 - x**3 - y**5) * np.exp(-x**2-y**2) \
      - 1/3*np.exp(-(x+1)**2 - y**2)
  return z
```

```python
# create the landscape
x = np.linspace(-3,3,201)
y = np.linspace(-3,3,201)

Z = peaks(x,y)

# let's have a look!
plt.imshow(Z,extent=[x[0],x[-1],y[0],y[-1]],vmin=-5,vmax=5,origin='lower')
plt.show()
```

```python
# create derivative functions using sympy

sx,sy = sym.symbols('sx,sy')

sZ = 3*(1-sx)**2 * sym.exp(-(sx**2) - (sy+1)**2) \
      - 10*(sx/5 - sx**3 - sy**5) * sym.exp(-sx**2-sy**2) \
      - 1/3*sym.exp(-(sx+1)**2 - sy**2)


# create functions from the sympy-computed derivatives
df_x = sym.lambdify( (sx,sy),sym.diff(sZ,sx),'sympy' )
df_y = sym.lambdify( (sx,sy),sym.diff(sZ,sy),'sympy' )

df_x(1,1).evalf()
```

```python
# random starting point (uniform between -2 and +2)
localmin = np.random.rand(2)*4-2 # also try specifying coordinates
startpnt = localmin[:] # make a copy, not re-assign

# learning parameters
learning_rate = .01
training_epochs = 1000

# run through training
trajectory = np.zeros((training_epochs,2))
for i in range(training_epochs):
  grad = np.array([ df_x(localmin[0],localmin[1]).evalf(), 
                    df_y(localmin[0],localmin[1]).evalf() 
                  ])
  localmin = localmin - learning_rate*grad  # add _ or [:] to change a variable in-place
  trajectory[i,:] = localmin


print(localmin)
print(startpnt)
```

```python
# let's have a look!
plt.imshow(Z,extent=[x[0],x[-1],y[0],y[-1]],vmin=-5,vmax=5,origin='lower')
plt.plot(startpnt[0],startpnt[1],'bs')
plt.plot(localmin[0],localmin[1],'ro')
plt.plot(trajectory[:,0],trajectory[:,1],'r')
plt.legend(['rnd start','local min'])
plt.colorbar()
plt.show()
```
